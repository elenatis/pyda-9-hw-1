### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
ratings = pd.read_csv('ml-latest/ratings.csv')
movies = pd.read_csv('ml-latest/movies.csv')
top_ratings = ratings[ratings['rating'] == 5][['rating', 'movieId']]
```

```python
top_movie = top_ratings.groupby('movieId').count().sort_values('rating').tail(1)
top_movie.merge(movies, on = 'movieId', how = 'left')
```

```python
top_movie_id = int(top_ratings.movieId.value_counts().index[0])
movies[movies['movieId'] == top_movie_id]
```

### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
```

```python
filtered_countries = data[(data.country.isin(['Latvia','Lithuania','Estonia'])) & (data.category.isin(['4','12','21']) ) & ((2005 <= filtered_countries.year) & (filtered_countries.year <= 2010 ))]

# проверка
print(filtered_countries.country.unique(), filtered_countries.category.unique(), filtered_countries.year.unique())

filtered_countries[filtered_countries.quantity > 0].quantity.sum()
```
